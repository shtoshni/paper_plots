```python
from os import path
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch


# Set Matplotlib Parameters
plt.style.use('bmh')
plt.rc('text', usetex=True)   #  Might require installing TeX fonts
plt.rc('axes', facecolor='w', labelcolor='b', edgecolor='k', labelsize=20)
plt.rcParams['font.size'] = 18
plt.rcParams['legend.fontsize'] = 18
plt.rcParams['xtick.labelsize'] = 18
plt.rcParams['ytick.labelsize'] = 18
```

```python
plt.figure(figsize=(6, 5))

plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)

questions = np.array([64_000, 128_000, 256_000, 512_000, 1_024_000])

gt_scores = np.array([41.35, 43.85, 44.95, 48.42, 49.62])
gt_std = np.array([0.32, 1.43, 1.00, 0.54, 0.59])

err_10 = np.array([40.05, 43.60, 45.95, 49.38, 49.77])
err_10_std = np.array([0.71, 0.65, 1.20, 0.58, 0.58])


err_20 = np.array([41.02, 43.38, 45.25, 49.17, 49.90])
err_20_std = np.array([0.48, 1.39, 0.3, 1.05, 0.58])

err_40 = np.array([38.40, 41.35, 44.08, 46.80, 48.12])
err_40_std = np.array([0.28, 1.03, 0.56, 0.14, 0.50])

err_80 = np.array([25.93, 28.77, 30.07, 31.57, 32.55])
err_80_std = np.array([0.86, 1.56, 0.99, 1.23, 1.07])

alpha = 0.2
plt.plot(questions, gt_scores, color='#1f77b4', linewidth=2.0, label='Original', marker='o')
plt.fill_between(questions, gt_scores - gt_std, gt_scores + gt_std, alpha=alpha, color='#1f77b4')

plt.plot(questions, err_10, linestyle='--', color='#2ca02c', linewidth=2.0, label='Incorrect 10\%', marker='s')
plt.fill_between(questions, err_10 - err_10_std, err_10 + err_10_std, alpha=alpha, color='#2ca02c')

plt.plot(questions, err_20, linestyle=':', color='#ff7f0e', linewidth=2.0, label='Incorrect 20\%', marker='^')
plt.fill_between(questions, err_20 - err_20_std, err_20 + err_20_std, alpha=alpha, color='#ff7f0e')

plt.plot(questions, err_40, linestyle='--', color='#9467bd', linewidth=2.0, label='Incorrect 40\%', marker='D')
plt.fill_between(questions, err_40 - err_40_std, err_40 + err_40_std, alpha=alpha, color='#9467bd')

plt.plot(questions, err_80, linestyle='-', color='#d62728', linewidth=2.0, label='Incorrect 80\%', marker='*')
plt.fill_between(questions, err_80 - err_80_std, err_80 + err_80_std, alpha=alpha, color='#d62728')


plt.xscale('log')
plt.ylim(13, 52)


plt.legend(loc='lower right', fontsize=16)

plt.xticks([64_000,  128_000, 256_000, 512_000, 1_024_000], ['64K', '128K', '256K', '512K', '1024K'])
plt.minorticks_off()


plt.xlabel('Size of SFT Dataset (in K)', color="grey", weight='black')
plt.ylabel('MATH Validation Accuracy (in \%)', color="grey", weight='black')


plt.tight_layout(pad=0.07)
plt.savefig("../plots/noise_pairing_with_err_bar.jpg", format='jpg', dpi=300)
plt.savefig("../plots/noise_pairing_with_err_bar.pdf", format='pdf', dpi=300)
plt.show()
```
