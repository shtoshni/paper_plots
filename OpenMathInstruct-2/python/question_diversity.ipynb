```python
from os import path
import numpy as np
import matplotlib
# matplotlib.use('Qt5Agg')  # or try 'Qt5Agg' or others

import matplotlib.pyplot as plt
```

```python
# Set Matplotlib Parameters
plt.style.use('bmh')
plt.rc('text', usetex=True)   #  Might require installing TeX fonts
plt.rc('axes', facecolor='w', labelcolor='b', edgecolor='k', labelsize=20)
plt.rcParams['font.size'] = 16
plt.rcParams['legend.fontsize'] = 16
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
```

```python
fig, ax = plt.subplots(figsize=(6, 5))  # A square figure with equal width and height
plt.grid(True, which='both', linestyle='--', color='#CCCCCC', linewidth=0.4)


# Data points
questions = np.array([1e3, 2e3, 4e3, 6445])
validation_scores = np.array([35.32, 40.12, 43.33, 46.50])
errors = np.array([0.86, 0.62, 0.51, 0.39])

alpha = 0.2
plt.plot(questions, validation_scores, color='#1f77b4', linewidth=2.0, label='Original', marker='o')
plt.fill_between(questions, validation_scores - errors, validation_scores + errors, alpha=alpha, color='#1f77b4')


plt.xscale('log')
plt.ylim(33, 48)

# Manually set the xticks for the logarithmic scale
plt.xticks([1e3, 2e3, 3e3, 4e3, 6e3], ['1K', '2K', '3K', '4K', '6K'])
plt.minorticks_off()

plt.xlabel('\# of Unique Questions', color="grey", weight='black')
plt.ylabel('MATH Validation Accuracy (in \%)', color="grey", weight='black')

plt.tight_layout(pad=0.07)
plt.savefig("../plots/question_diversity.jpg", format='jpg', dpi=300)
plt.savefig("../plots/question_diversity.pdf", format='pdf', dpi=300)
plt.show()
```
